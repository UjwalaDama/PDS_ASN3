```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
diabetes_data = pd.read_csv('/content/diabetes (1).csv')
np.random.seed(49)
sample_data = diabetes_data.sample(n=25)
population_bmi_98th_percentile = np.percentile(diabetes_data['BMI'], 98)
sample_bmi_98th_percentile = np.percentile(sample_data['BMI'], 98)
print("Population 98th Percentile BMI:", population_bmi_98th_percentile)
print("Sample 98th Percentile BMI:", sample_bmi_98th_percentile)
plt.figure(figsize=(8, 6))
plt.hist(diabetes_data['BMI'], bins=30, alpha=0.5, label='Population BMI')
plt.hist(sample_data['BMI'], bins=10, alpha=0.5, label='Sample BMI')
plt.axvline(population_bmi_98th_percentile, color='red', linestyle='dotted', linewidth=2, label='Population 98th Percentile')
plt.axvline(sample_bmi_98th_percentile, color='yellow', linestyle='dotted', linewidth=2, label='Sample 98th Percentile')
plt.title('BMI Comparison: Population vs Sample')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.legend()
plt.show()

```
